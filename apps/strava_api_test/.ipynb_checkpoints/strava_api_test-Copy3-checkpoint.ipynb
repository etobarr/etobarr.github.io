```python
# import packages
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import numpy as np
import pandas as pd
pd.set_option("display.max.columns", None)
import requests
import json
import urllib3
import matplotlib.pyplot as plt # import the scripting layer
import seaborn as sns
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# refresh access token (expires every 6 hours)
auth_url = "https://www.strava.com/oauth/token"
activities_url = "https://www.strava.com/api/v3/athlete/activities"

payload = {
    'client_id': "116573",
    'client_secret': 'f178359239fbb2c061697c8163a874471ca263de',
    'refresh_token': 'c00d945cd19b4238d9bdb9a5ebfcc1cff48ec05a',
    'grant_type': "refresh_token",
    'f': 'json'
}

print("Requesting Token...\n")
res = requests.post(auth_url, data=payload, verify=False)
access_token = res.json()['access_token']
print("Access Token = {}\n".format(access_token))

# setup
header = {'Authorization': 'Bearer ' + access_token}
request_page_num = 1
all_activities = []
```

```python
# setup
header = {'Authorization': 'Bearer ' + access_token}
request_page_num = 1
activities_byPage = []

while True:
    # fetches data from api; limited to 200 activities per request, must retrieve by page
    param = {'per_page': 200, 'page': request_page_num}
    page_data = []
    page_data = requests.get(activities_url, headers=header, params=param).json()

    # break loop if page has no activities
    if len(page_data) == 0:
        print("retrieved all activities :)")
        break
        
    # break loop if exceeeded api rate limit
    if page_data[0].get('message') == 'Rate Limit Exceeded':
        print('Rate Limit Exceeded :(')
        break
    
    # build list of activities from pages and un-nest activities from pages
    print('retrieved ' + str(len(page_data)) + ' activities from page ' + str(request_page_num))
    activities_byPage.append(page_data)
    all_activities = [item for sublist in activities_byPage for item in sublist]
    
    request_page_num += 1

if len(all_activities) > 0:
    # write activities to .json file
    with open('strava_data.json', 'w') as data_file_write:
        json.dump(all_activities, data_file_write, indent=4)
    print('data file created/overwritten')
```

```python
# WARNING: this will reset loaded data
# load stored .json data to list and convert it to a pd dataframe
with open('strava_data.json', 'r') as data_file_read:
    all_activities_read = json.load(data_file_read)
activities_df = pd.DataFrame(all_activities_read)

# filter to only cycling activities
rides_df = activities_df.copy()
rides_df = activities_df[(activities_df['type'].isin(['Ride', 'VirtualRide']))]

# convert 'device watts' column to boolean and filter to only cycling activities with power data
rides_withPower_df = rides_df.copy()
rides_withPower_df['device_watts'] = pd.to_numeric(rides_withPower_df['device_watts'], errors='coerce')
rides_withPower_df['device_watts'].fillna(False, inplace=True)
rides_withPower_df['device_watts'] = rides_withPower_df['device_watts'].astype(bool)
rides_withPower_df = rides_withPower_df[rides_withPower_df['device_watts']]

# create columns
metrics = ['watts', 'time']
for metric in metrics:
    rides_withPower_df[f'{metric}Stream'] = None
rides_withPower_df.head()
```

```python
rides_withPower_df = pd.read_json('strava_data.json', orient='records', lines=True)
```

```python
count = 0
for index, row in rides_withPower_df.iterrows():
    for metric in metrics:
        if rides_withPower_df.at[index, f'{metric}Stream'] == None:
            # define parameters for request
            stream_params = {'keys': {metric}, 'key_by_type': 'true'}

            # request stream data
            stream_url = f"https://www.strava.com/api/v3/activities/{row['id']}/streams"
            stream_data = requests.get(stream_url, headers=header, params=stream_params).json()
            
            if stream_data.get('message') == 'Rate Limit Exceeded':
                print('Rate Limit Exceeded :(')
                break
            
            # assign power data to the appropriate ride in the DataFrame
            try:
                rides_withPower_df.at[index, f'{metric}Stream'] = stream_data[f'{metric}']['data']
                count += 1
            except KeyError:
                rides_withPower_df.at[index, f'{metric}Stream'] = 0
        else:
            continue
    if stream_data.get('message') == 'Rate Limit Exceeded':
                print('Rate Limit Exceeded :(')
                break

if count > 0:
    print(str(count) + ' cells updated with stream data :)')
rides_withStreamData_df = rides_withPower_df.copy()
rides_withStreamData_df.head()
```
